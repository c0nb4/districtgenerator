```python
# Full Simlation Worklow for the Districtgenerator 
# 
from districtgenerator.classes import datahandler 
import pandas as pd
import os


current_dir = globals()['_dh'][0]
```

```python
# Load the CSV file
precleaned_data = pd.read_csv('20250120_Mierendorff_Scenario.csv', sep=';')

# Set the 'rettofit' column to 0
precleaned_data['retrofit'] = 0
precleaned_data['construction_type'] = ""
precleaned_data['night_setback'] = 0
precleaned_data['heater'] = "HP"
precleaned_data['PV'] = 0
precleaned_data['STC'] = 0
precleaned_data['EV'] = 0 
precleaned_data['BAT'] = 0
precleaned_data['f_TES'] = 0
precleaned_data['f_BAT'] = 0
precleaned_data['f_EV'] = "M"
precleaned_data['f_PV'] = 0
precleaned_data['f_STC'] = 0
precleaned_data['gamma_PV'] = 0
precleaned_data['ev_charging'] = "on_demand"

# Drop all buildings wheare area & year is nan

clean_data = precleaned_data.dropna(subset = ['area', 'year', 'building'])
clean_data = clean_data[clean_data["area"] != 0]
clean_data = clean_data[clean_data["building"] != '-']

clean_data.loc[(clean_data["year"] > 1979) & (clean_data["building"] == "AB"), "building"] = "MFH"
clean_data['year'] = clean_data['year'].astype(int)

clean_data.info()
```

```python

# Save the updated DataFrame back to the CSV file
clean_data.reset_index(drop = True, inplace = True)
clean_data.loc[:,"id"]    = clean_data.index
clean_data.to_csv('districtgenerator/data/scenarios/20252001_Mierendorff_No_Retrofit.csv', index=False, sep=';')
```
