```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os


```

```python
for file in os.listdir("test_scenario"):
    if file.endswith(".csv"):
        df = pd.read_csv(f"test_scenario/{file}", sep=",")
        # Only plot heat columns
        heat_cols = [col for col in df.columns if col == "heat"]
        if heat_cols:
            df[heat_cols].plot(title=file)
```
